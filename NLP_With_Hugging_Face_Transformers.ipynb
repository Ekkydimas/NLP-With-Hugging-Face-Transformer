```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification")
text = "Transformers library has made NLP accessible to everyone."
candidate_labels = ["technology", "education", "science", "entertainment", "business"]
print("Zero-shot Classification:")
print(classifier(text, candidate_labels=candidate_labels))
```

```python
from transformers import pipeline
generator = pipeline("text-generation", model="gpt2")
seed_text = "Artificial Intelligence is transforming"
generated_texts = generator(seed_text, max_length=50, num_return_sequences=3, temperature=0.8)
print("\nText Generation:")
for i, gen in enumerate(generated_texts):
    print(f"Generated Text {i+1}: {gen['generated_text']}")
```

```python
from transformers import pipeline

# Inisialisasi pipeline fill-mask
unmasker = pipeline("fill-mask", model="bert-base-uncased")

# Gunakan token [MASK] yang sesuai
masked_sentence = "The [MASK] technology is advancing rapidly."

# Prediksi untuk kata yang di-mask
top_predictions = unmasker(masked_sentence, top_k=3)

# Cetak hasil prediksi
print("\nFill-Mask Predictions:")
for pred in top_predictions:
    print(f"{pred['sequence']} (Score: {pred['score']:.4f})")
```

```python
from transformers import pipeline
ner = pipeline("ner", grouped_entities=True)
sentence = "Hugging Face was founded in 2016 in New York."
entities = ner(sentence)
print("\nNamed Entity Recognition:")
for entity in entities:
    print(f"Entity: {entity['word']}, Type: {entity['entity_group']}, Confidence: {entity['score']:.4f}")
```

```python
from transformers import pipeline
qa_pipeline = pipeline("question-answering")
context = "Natural Language Processing is a field of AI focused on language understanding and generation."
question = "What is NLP?"
answer = qa_pipeline(question=question, context=context)
print("\nQuestion Answering:")
print(f"Question: {question}")
print(f"Answer: {answer['answer']}")
```

```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis")
sentences = [
    "I love how transformers make NLP easier.",
    "The library sometimes feels overwhelming.",
    "Overall, it's an amazing tool."
]
print("\nSentiment Analysis:")
for sentence in sentences:
    result = classifier(sentence)
    print(f"Sentence: {sentence} -> Sentiment: {result[0]['label']} (Confidence: {result[0]['score']:.4f})")
```

```python
from transformers import pipeline

translator = pipeline("translation_id_to_en", model="Helsinki-NLP/opus-mt-id-en")

text_to_translate = "Hari ini aku kehujanan"
result = translator(text_to_translate)

print(result[0]['translation_text'])
```

# Analisis

Analisis Kode NLP dengan Hugging Face Transformers
Dalam kode ini, saya menggunakan berbagai pipeline dari Hugging Face Transformers untuk menyelesaikan beberapa tugas Natural Language Processing (NLP) yang berbeda. Salah satu bagian yang paling menarik adalah Zero-Shot Classification. Metode ini memungkinkan kita untuk mengklasifikasikan teks ke dalam kategori tertentu tanpa perlu melatih model terlebih dahulu pada dataset berlabel. Hal ini menjadi sangat praktis dan efisien karena hanya membutuhkan input teks dan label kandidat. Pada contoh kode, kalimat "Transformers library has made NLP accessible to everyone" diklasifikasikan ke dalam kategori seperti "technology", "education", "science", dan "business", di mana model memberikan probabilitas untuk setiap label sesuai relevansinya.

Selain itu, kode ini juga menerapkan Text Generation menggunakan model GPT-2. Dengan teks awal "Artificial Intelligence is transforming", model menghasilkan teks lanjutan dengan panjang maksimum 50 kata dan variasi hasil yang diatur melalui parameter temperature. Ini menunjukkan kemampuan model dalam membuat teks kreatif yang masih relevan dengan input awal. Selanjutnya, terdapat Fill-Mask menggunakan model BERT, yang berguna untuk memprediksi kata yang hilang dalam kalimat seperti "The [MASK] technology is advancing rapidly". Model berhasil memberikan prediksi kata seperti "AI" atau "new" sebagai kandidat terbaik.

Kemudian, pipeline Named Entity Recognition (NER) digunakan untuk mengenali entitas seperti nama, lokasi, dan organisasi dalam kalimat "Hugging Face was founded in 2016 in New York". Model berhasil mengidentifikasi entitas "Hugging Face" sebagai organisasi, "2016" sebagai waktu, dan "New York" sebagai lokasi, dengan tingkat kepercayaan yang tinggi. Sementara itu, pada bagian Question Answering, model menjawab pertanyaan berdasarkan konteks yang diberikan. Dengan input "Apa itu NLP?" dan konteks tentang Natural Language Processing, model memberikan jawaban yang tepat dan akurat.

Selanjutnya, kode ini juga menerapkan Sentiment Analysis untuk menganalisis opini dari beberapa kalimat. Model mampu mengidentifikasi sentimen positif pada kalimat seperti "I love how transformers make NLP easier" dan memberikan confidence score yang relevan. Di bagian Summarization, pipeline digunakan untuk merangkum teks panjang mengenai perubahan pendidikan teknik di Amerika. Hasil ringkasan menunjukkan kemampuan model dalam merangkum informasi kompleks menjadi lebih ringkas dan mudah dipahami.

Terakhir, pipeline Translation digunakan untuk menerjemahkan teks dari bahasa Indonesia ke bahasa Inggris menggunakan model "Helsinki-NLP/opus-mt-id-en". Dengan input "Hari ini aku kehujanan", model berhasil menerjemahkannya menjadi "Today I got rained on". Ini menunjukkan fleksibilitas model dalam menerjemahkan teks lintas bahasa secara cepat dan akurat.

Secara keseluruhan, kode ini mencakup berbagai aspek penting dalam NLP, seperti klasifikasi, generasi teks, tanya jawab, analisis sentimen, penerjemahan, dan pengisian teks. Penggunaan pipeline dari Hugging Face memudahkan implementasi model pre-trained tanpa perlu melakukan pelatihan tambahan, sehingga sangat praktis dan efisien untuk berbagai tugas NLP.
